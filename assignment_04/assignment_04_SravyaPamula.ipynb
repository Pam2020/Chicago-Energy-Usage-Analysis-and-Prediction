### Question 8 and 9: 

We look at the data dictionary  and find what these 9 chosen variables mean.

1. UNITID - A unique identification number assigned to postsecondary institutions through the IPEDS survey.

2. INSTNM - The institution name as reported by IPEDS. 

3. CITY - The city where the institute is located as reported by IPEDS. 

4. STABBR - The state where the institute is located as reported by IPEDS. 

5. ZIP -  Institute zipcode. 

6. ADM_RATE -  This is the Fall admissions rate at that campus defined as the number of undergraduates admitted by the number of undergraduate admissions. 

7. UGDS - The number of degree/certificate seeking undergraduate students enrolled in the institute. 

8. TUITIONFEE_IN - Tuition and required fees for in-state students in the given institute. 

9. MN_EARN_WNE_P10 - Mean earnings of graduates working 10 years after the entry. This column seems to have only NaN values.
